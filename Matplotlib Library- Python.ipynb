```python
import matplotlib.pyplot as plt
```

```python
# Linear Graph
```

```python
x=[1,2,3,4]
y=[5,6,7,8]
plt.plot(x,y)
plt.show()
```

```python
x=[1,2,3,4]
y=[5,6,7,8]
c='r'                                       #Change the color of the graph
plt.plot(x,y,c)
plt.show()
```

```python
x=[5,7,2,8]
y=[2,9,4,6]
plt.plot(x,y)
plt.show()
```

```python
x=[5,7,2,8]
y=[2,9,4,6]
plt.plot(x,y,marker='o')
plt.show()
```

```python
x=[5,7,2,8]
y=[2,9,4,6]
plt.plot(x,y)
plt.plot(y,marker='>')                   # Put marker on y-axis coordinates
plt.show()
```

```python
x=[5,7,2,8]
y=[2,9,4,6]
plt.plot(x,y)
plt.plot(y,marker='>')                   # Put marker on y-axis coordinates
plt.plot(x,marker='o')                   # Put marker on x-axis coordinates  
plt.show()
```

```python
# BAR graph
```

```python
x=[1,2,3,4]
y=[5,6,7,8]
plt.bar(x,y)
plt.show()
```

```python
x=[1,2,3,4]
y=[5,6,7,8]
c=['r','g','b','y']
plt.bar(x,y, color=c)
plt.show()
```

```python
import numpy as np
```

```python
x=np.array(['A','B','C','D'])
y=np.array([2,4,6,8])
plt.barh(x,y)                     #barh represent horizontal graph
plt.show()
```

```python
x=np.array(['A','B','C','D'])
y=np.array([2,4,6,8])
plt.barh(x,y,color='y')
plt.show()
```

```python
# change the width of the graph
```

```python
x=np.array(['A','B','C','D'])
y=np.array([2,4,6,8])
plt.bar(x,y,color='pink',width=0.2)
plt.show()
```

```python
x=np.array(['A','B','C','D'])
y=np.array([2,4,6,8])
plt.barh(x,y,color='skyblue',height=0.1)
plt.show()
```

```python
# Scatter Plot
```

```python
x=[1,2,3,4,5]
y=[6,7,8,9,10]
plt.title(' Scatter Graph')
plt.xlabel('months')
plt.ylabel('number')
plt.scatter(x,y)
plt.show()
```

```python
x=[1,2,3,4,5]
y=[6,7,8,9,10]
plt.title(' Scatter Graph')
plt.xlabel('months')
plt.ylabel('number')
c=['red','magenta','pink','yellow','black']
plt.scatter(x,y,color=c)                        # change the color
plt.show()
```

```python
x=[1,2,3,4,5]
y=[6,7,8,9,10]
plt.title(' Scatter Graph')
plt.xlabel('months')
plt.ylabel('number')
c=['red','magenta','pink','yellow','black']
plt.scatter(x,y,color=c,s=150 )                        # change the size
plt.show()
```
