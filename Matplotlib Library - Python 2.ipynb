```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

fname=r'bird.jpg'

# openning image using PIL

image=Image.open(fname).convert('RGB')

# Mapping image to gray image

plt.imshow(image,cmap='spring')
plt.title('BIRD')
plt.show()
```

```python
x=[10,20,30,40]
y=['english','hindi','maths','science']

plt.pie(x, labels=y)
plt.show()
```

```python
x=[10,20,30,40]
y=['english','hindi','maths','science']
c=['yellow','pink','magenta','blue']                #change color of the pie chart
plt.pie(x, labels=y,colors=c)
plt.show()
```

```python
x=[10,20,30,40]
y=['english','hindi','maths','science']

plt.pie(x, labels=y)
plt.legend()
plt.show()
```
